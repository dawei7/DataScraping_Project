```python
import networkx as nx
import matplotlib.pyplot as plt
```

```python
# Read in dict_network.txt
dic = ""
with open(r'dict_network.txt','r') as f:
    for i in f.readlines():
        dic+=i #string
dict_network = eval(dic) # this is orignal dict with instace dict
```

```python

def show_actors_network(edge_weight=None):
    best_paid_actors=[
    ("Dwayne Johnson","https://www.imdb.com/name/nm0425005/"), #Dwayne Johnson
    ("Robert Downey Jr.","https://www.imdb.com/name/nm0000375/"), #Robert Downey Jr.
    ("Chris Hemsworth","https://www.imdb.com/name/nm1165110/"), #Chris Hemsworth
    ("Akshay Kumar","https://www.imdb.com/name/nm4289673/"), #Akshay Kumar
    ("Jackie Chan","https://www.imdb.com/name/nm0000329/"), #Jackie Chan
    ("Bradley Cooper","https://www.imdb.com/name/nm0177896/"), #Bradley Cooper
    ("Adam Sandler","https://www.imdb.com/name/nm0001191/"), #Adam Sandler
    ("Scarlett Johansson","https://www.imdb.com/name/nm0424060/"), #Scarlett Johansson
    ("Sofía Vergara","https://www.imdb.com/name/nm0005527/"), #Sofía Vergara
    ("Chris Evans","https://www.imdb.com/name/nm0262635/") #Chris Evans
    ]

    G = nx.Graph()
    G.add_nodes_from(actor[0] for actor in best_paid_actors)


    edges_level_one = []
    edges_level_two = []
    edges_level_three = []

    for k,v in dict_network.items():
        if v != None:
            if v==1 and (edge_weight == None or edge_weight == 1):
                color = "green" 
                edges_level_one.append(k)
                G.add_edge(*k)
                G.add_edge(*k,weight=v)
            if v==2 and (edge_weight == None or edge_weight == 2):
                color = "blue"
                edges_level_two.append(k)
                G.add_edge(*k)
                G.add_edge(*k,weight=v)
            if v==3 and (edge_weight == None or edge_weight == 3):
                color = "red"
                edges_level_three.append(k)
                G.add_edge(*k)
                G.add_edge(*k,weight=v)


    pos = nx.circular_layout(G)
    weights = nx.get_edge_attributes(G,'weight')
    colors = nx.get_edge_attributes(G,'color')
    nx.draw(G,pos,with_labels = True,node_color='lime',node_size=2000)

    nx.draw_networkx_edges(G,pos,edgelist=edges_level_one,width=3,edge_color="green")
    nx.draw_networkx_edges(G,pos,edgelist=edges_level_two,width=3,edge_color="blue")
    nx.draw_networkx_edges(G,pos,edgelist=edges_level_three,width=3,edge_color="red")

    nx.draw_networkx_edge_labels(G,pos,edge_labels=weights)
    plt.rcParams['figure.figsize'] = [10, 10]
    #plt.subplot(2,2,subplot_pos)
    network_level = "Full" if edge_weight == None else edge_weight
    plt.title(f"Network Level {network_level}")
    plt.show(G,pos)
```

```python
show_actors_network()
```

```python
show_actors_network(edge_weight=1)
```
