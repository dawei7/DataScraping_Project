```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud, STOPWORDS, ImageColorGenerator
```

```python
def get_wordcloud(actor):
    dic = ""
    with open(f'dict_{actor}_mov_descr.txt','r') as f:
        for i in f.readlines():
            dic+=i #string
        dict_actor_mov_descr = eval(dic) # this is orignal dict with instace dict

    unwanted_chars = ".,-_" #Filter unwanted chars out

    words = []
    for raw_words in dict_actor_mov_descr.values(): #Loop through dict
        for raw_word in raw_words.split(" "): #Loop through splitted words
            words.append(raw_word.strip(unwanted_chars)) #append all words to a list

    stopwords = set(STOPWORDS) #get stopwords from library
    wordcloud = WordCloud(stopwords=stopwords).generate(" ".join(words)) #Join as " " separated string for wordcloud function

    # Display the generated image:
    plt.imshow(wordcloud, interpolation='bilinear') #Generate Wordcloud
    plt.axis("off")
    plt.show()
```
